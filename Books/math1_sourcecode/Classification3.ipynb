```python
import numpy as np
import matplotlib.pyplot as plt
```

データの入力

```python
# 学習データを読み込む
train = np.loadtxt('./data3.csv', delimiter=',', skiprows=1)
train_x = train[:,0:2]
train_y = train[:,2]
```

```python
# パラメータを初期化
theta = np.random.rand(4)

# 標準化
mu = train_x.mean(axis=0)
sigma = train_x.std(axis=0)
def standardize(x):
    return (x - mu) / sigma

train_z = standardize(train_x)
```

```python
# プロットして確認
plt.plot(train_z[train_y == 1, 0], train_z[train_y == 1, 1], 'o')
plt.plot(train_z[train_y == 0, 0], train_z[train_y == 0, 1], 'x')
plt.show()
```

```python
# x0とx3を加える
def to_matrix(x):
    x0 = np.ones([x.shape[0], 1])
    x3 = x[:,0,np.newaxis] ** 2
    return np.hstack([x0, x, x3])

X = to_matrix(train_z)
```

$
    f_\theta(x) = \frac{1}{(1+ \exp(-\theta^{T}x)}
$

```python
# シグモイド関数
def f(x):
    return 1 / (1 + np.exp(-np.dot(x, theta)))
```

$
    f(x) = \left\{ \begin{array}{l}
    \displaystyle 1 \;\;\; (f_\theta(x) \geq 0.5) \\
    0 \;\;\; (f_\theta(x) < 0.5) 
    \end{array} \right.
$

```python
# 分類関数
def classify(x):
    return (f(x) >= 0.5).astype(np.int)
```

```python
# 学習率
ETA = 1e-3

# 繰り返し回数
epoch = 5000

# 更新回数
count = 0
```

```python
# 学習を繰り返す
for _ in range(epoch):
    theta = theta - ETA * np.dot(f(X) - train_y, X)

    # ログの出力
    count += 1
    print('{}回目: theta = {}'.format(count, theta))
```

```python
# プロットして確認
x1 = np.linspace(-2, 2, 100)
x2 = -(theta[0] + theta[1] * x1 + theta[3] * x1 ** 2) / theta[2]
plt.plot(train_z[train_y == 1, 0], train_z[train_y == 1, 1], 'o')
plt.plot(train_z[train_y == 0, 0], train_z[train_y == 0, 1], 'x')
plt.plot(x1, x2, linestyle='dashed')
plt.show()
```

精度の評価

$
    Accuracy = \frac{TP + TN}{(TP + FP + FN + TN)}
$

```python
theta = np.random.rand(4)

accuracies = []

# 学習を繰り返す
for _ in range(epoch):
    theta = theta - ETA * np.dot(f(X) - train_y, X)
    
    result = classify(X) == train_y
    accuracy = len(result[result == True])/len(result)
    accuracies.append(accuracy)
    
x = np.arange(len(accuracies))

plt.plot(x, accuracies)
plt.show()
```

確率的勾配降下法（ミニバッジ法）

$
    \theta_j := \theta_j - \eta \sum_{k \in K} (f_\theta(x^{(k)}) - y^{(k)})x_{j}^{(k)}
$

```python
theta = np.random.rand(4)

for _ in range(epoch):
    p = np.random.permutation(X.shape[0])
    for x, y in zip(X[p, :], train_y[p]):
        theta = theta - ETA * (f(x) - y)*x
        
```

```python
x1 = np.linspace(-2, 2, 100)
x2 = -(theta[0] + theta[1] * x1 + theta[3] *x1 **2)/theta[2]

plt.plot(train_z[train_y == 1, 0], train_z[train_y == 1, 1], 'o')
plt.plot(train_z[train_y == 0, 0], train_z[train_y == 0, 1], 'x')
plt.plot(x1, x2, linestyle = 'dashed')
plt.show()
```
