Q.7.1  
★★ 以下の折れ線グラフを1行2列に配置して描き、変数名をそれぞれのタイトルとして表示せよ。  
また、subplots()を用いよ。  
X = [1, 2, 3, 4, 5]  # 横軸

Q.7.2  
★★ 以下の折れ線グラフを1行2列に配置して描き、変数名をそれぞれのタイトルとして表示せよ。  
また、add_subplot()を用いよ。  
X = [1, 2, 3, 4, 5] # 横軸  
food = [23, 21, 42, 32, 23]  
drink = [33, 25, 41, 36, 16]

```python
X = [1, 2, 3, 4, 5] # 横軸
food = [23, 21, 42, 32, 23]
drink = [33, 25, 41, 36, 16]

fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(X, food)
ax1.set_title("food")

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(X, drink)
ax2.set_title("drink")

plt.show()
```

Q.7.3  
★★以下の棒グラフを2行1列で描き、横軸のティックラベルを指定せよ。  
その際、tight_layoutで図が重ならないようにせよ。  
また、変数名をy軸のラベルに設定せよ。  
tick_label = ["A", "B", "C", "D", "E"]  
food = [23, 21, 42, 32, 23]  
drink = [33, 25, 41, 36, 16]  
ヒント:axにラベルを指定→ax.set_ylabel(args)

```python
tick_label = ["A", "B", "C", "D", "E"]
food = [23, 21, 42, 32, 23]
drink = [33, 25, 41, 36, 16]

fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)
ax1.bar(tick_label, food)
ax1.set_ylabel("food")

ax2 = fig.add_subplot(2, 1, 2)
ax2.bar(tick_label, drink)
ax2.set_ylabel("drink")

plt.tight_layout()

plt.show()
```

Q.8.1  
★★ 以下の折れ線グラフを1行2列に配置して描き、変数名をそれぞれのタイトルとして表示せよ。  
その際、y軸のレンジを統一せよ。  
X = [1, 2, 3, 4, 5] # 横軸  
food = [23, 21, 42, 32, 23]  
drink = [33, 25, 41, 36, 16]

```python
X = [1, 2, 3, 4, 5] # 横軸
food = [23, 21, 42, 32, 23]
drink = [33, 25, 41, 36, 16]

fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(X, food)
ax1.set_title("food")
ymin, ymax = ax1.get_ylim()

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(X, drink)
ax2.set_title("drink")
ax2.set_ylim(ymin, ymax)

plt.show()
```

Q.9.1  
★★ 以下のデータを２軸グラフで表示せよ。また、変数名を凡例として表示せよ。  
sales = [293, 221, 240, 145, 223]  # 棒グラフ, 青, アルファ=0.5, ティックラベル=tick_label  
tick_label = ["week1", "week2", "week3", "week4", "week5"]  
temperature = [33, 28, 30, 24, 28] # 折れ線グラフ, 緑, マーカーx

```python
sales = [293, 221, 240, 145, 223]  # 棒グラフ, 青, アルファ=0.5, ティックラベル=tick_label  
tick_label = ["week1", "week2", "week3", "week4", "week5"]  
temperature = [33, 28, 30, 24, 28] # 折れ線グラフ, 緑, マーカーx

fig = plt.figure()

ax1 = fig.add_subplot(1, 1, 1)
ax1.bar(tick_label, sales, label="sales", color="b", alpha=0.5)

ax2 = ax1.twinx()
ax2.plot(temperature, label="temperature", color="r", marker="x")

handler1, label1 = ax1.get_legend_handles_labels()
handler2, label2 = ax2.get_legend_handles_labels()
plt.legend(handler1+handler2, label1+label2)

plt.show()
```
